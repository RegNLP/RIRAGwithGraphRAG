```python
import json
import os
import networkx as nx

# Initialize the graph
G = nx.DiGraph()



# ---- Graph Construction ----
for doc in documents:
    doc_id = doc["DocumentID"]
    doc_node_id = f"D{doc_id}"
    print(f"\n🟢 Processing Document {doc['title']} (ID: {doc_id})")

    G.add_node(doc_node_id, type="Document", title=doc["title"], version=doc["version"], date=doc["date"])
    #print(f"➕ Added Document node: {doc_node_id}")

    json_path = doc["json_file_path"]
    if not os.path.exists(json_path):
        print(f"  ❌ File not found: {json_path}")
        continue

    try:
        with open(json_path, "r", encoding="utf-8") as f:
            passages = json.load(f)
    except Exception as e:
        print(f"  ❌ Error loading JSON: {e}")
        continue

    passages_by_id = {}

    for p in passages:
        raw_passage_id = p.get("PassageID") or p.get("ID") or p.get("ContextID")
        if not raw_passage_id:
            continue

        passage_id = str(raw_passage_id).replace("..", ".").strip()
        if passage_id.endswith('.'):
            passage_id = passage_id[:-1]

        uid = p.get("ID") or p.get("ContextID")
        if not uid:
            continue

        text = p.get("Text") or p.get("Passage") or ""
        if not text.strip():
            text = "[PLACEHOLDER]"

        G.add_node(
            uid,
            type="Passage",
            text=text,
            passage_id=passage_id,
            document_id=doc_id
        )
        #print(f"    ➕ Passage node: {passage_id}")

        passages_by_id[passage_id] = uid

        # Named Entities
        named_entities = p.get("NamedEntities", [])
        for entity in named_entities:
            if isinstance(entity, dict):
                term = entity.get('ContextID') or entity.get('ID') or entity.get('Term', '')
                description = entity.get('Description') or entity.get('Meaning', '')
            elif isinstance(entity, str):
                term = entity
                description = ''
            else:
                #print(f"    ⚠️ Skipping invalid NamedEntity format in passage {passage_id}: {entity}")
                continue

            if not term:
                #print(f"    ⚠️ Skipping empty NamedEntity in passage {passage_id}")
                continue

            ne_uid = f"NE_{term}_{uid}"  # Still use uid internally for uniqueness
            G.add_node(ne_uid, type="NamedEntity", term=term, description=description)
            G.add_edge(uid, ne_uid, type="MENTIONS")
            print(f"      🟡 NamedEntity node added: {term}")
            print(f"      🔗 MENTIONS edge: {passage_id} → {term}")

        # Defined Terms
        defined_terms = p.get("DefinedTerms", [])
        for term in defined_terms:
            if isinstance(term, dict):
                t = term.get('ContextID') or term.get('ID') or term.get('Term', '')
                desc = term.get('Description') or term.get('Meaning', '')
            elif isinstance(term, str):
                t = term
                desc = ''
            else:
                #print(f"    ⚠️ Skipping invalid DefinedTerm format in passage {passage_id}: {term}")
                continue

            if not t:
                #print(f"    ⚠️ Skipping empty DefinedTerm in passage {passage_id}")
                continue

            dt_uid = f"DT_{t}_{uid}"  # Still use uid internally for uniqueness
            G.add_node(dt_uid, type="DefinedTerm", term=t, description=desc)
            G.add_edge(uid, dt_uid, type="USES_TERM")
            print(f"      🔵 DefinedTerm node added: {t}")
            print(f"      🔗 USES_TERM edge: {passage_id} → {t}")

    # ---- Link Passages Hierarchically or to Document ----
    for passage_id, uid in passages_by_id.items():
        current_passage_id = passage_id
        parent_found = False

        while not parent_found and '.' in current_passage_id:
            parent_passage_id = '.'.join(current_passage_id.split('.')[:-1])
            parent_uid = passages_by_id.get(parent_passage_id)

            if parent_uid:
                G.add_edge(parent_uid, uid, type="PARENT_OF")
                parent_found = True
                #print(f"    🔗 PARENT_OF: {parent_passage_id} → {passage_id}")
            else:
                print(f"    🔍 Parent not found for passage {passage_id}, moving to higher level: {parent_passage_id}")
                current_passage_id = parent_passage_id

        if not parent_found:
            G.add_edge(doc_node_id, uid, type="CONTAINS")
            print(f"  ⚠️ No parent or grandparent found for passage {passage_id}. Linking to document.")
```

```python
import pandas as pd

# Load the cross-reference data
try:
    crossref_df = pd.read_csv(crossref_path)
except Exception as e:
    print(f"❌ Failed to load cross-reference file: {e}")
    crossref_df = pd.DataFrame()

# Track number of added edges
added_edges = 0

for idx, row in crossref_df.iterrows():
    src_uid = str(row.get("SourceID"))
    tgt_uid = str(row.get("TargetID"))
    ref_text = row.get("ReferenceText", "").strip()
    ref_type = row.get("ReferenceType", "").strip()

    if src_uid in G.nodes and tgt_uid in G.nodes:
        # Add both directional edges
        G.add_edge(src_uid, tgt_uid, type="CITES", reference_text=ref_text, reference_type=ref_type)
        G.add_edge(tgt_uid, src_uid, type="CITED_BY", reference_text=ref_text, reference_type=ref_type)
        added_edges += 2

print(f"\n✅ Added {added_edges} cross-reference edges (CITES / CITED_BY)")
```

```python
import os
import json
import random
import numpy as np
import torch
from datasets import Dataset
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from transformers import (
    AutoModelForSequenceClassification,
    AutoTokenizer,
    Trainer,
    TrainingArguments,
    EarlyStoppingCallback,
)

# --- 1. Initialization and Configuration ---

print("--- Initializing Configuration ---")

# Define base paths for data and for saving models
# Ensure the drive is mounted in your Colab environment if using this path
data_folder = "/content/drive/MyDrive/Colab Notebooks/modeldatalocation"
models_output_folder = os.path.join(data_folder, "fine_tuned_models")
results_output_path = os.path.join(data_folder, "evaluation_results.json")

# Create the directory for saving models if it doesn't exist
os.makedirs(models_output_folder, exist_ok=True)

# Define the dictionary of models to process
# Mapping a short name to its Hugging Face checkpoint path
models_to_process = {
    "MiniLM": "sentence-transformers/all-MiniLM-L6-v2",
    "MPNet": "sentence-transformers/all-mpnet-base-v2",
    "MSMarco": "microsoft/MiniLM-L12-H384-uncased",
    "BERT": "bert-base-uncased"
}

# Define paths to the training and validation data
train_path = os.path.join(data_folder, "ObliQA_MultiPassage_train.json")
val_path = os.path.join(data_folder, "ObliQA_MultiPassage_val.json")

# Set the device to CUDA if available, otherwise CPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# --- 2. Data Preparation ---

def load_data_with_negatives(file_path: str) -> Dataset:
    """
    Reads a JSON file with questions and a list of relevant passages,
    generates negative examples, and returns a balanced, shuffled
    Hugging Face Dataset.

    Args:
        file_path (str): Path to the input JSON file with the new format.

    Returns:
        Dataset: A Hugging Face Dataset containing 'question', 'passage', and 'label' columns.
    """
    print(f"Loading and processing data from: {file_path}")
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}. Please check the path.")
        return None
    except json.JSONDecodeError:
        print(f"Error: Could not decode JSON from {file_path}. Please check file format.")
        return None

    # Extract all passages from the entire dataset for negative sampling
    all_passages = []
    for item in data:
        for passage_info in item.get('Passages', []):
            all_passages.append(passage_info['Passage'])

    if not all_passages:
        print("Warning: No passages found in the data file.")
        return Dataset.from_dict({'question': [], 'passage': [], 'label': []})

    processed_examples = []
    for item in data:
        question = item.get('Question')
        positive_passages = [p['Passage'] for p in item.get('Passages', [])]

        if not question or not positive_passages:
            continue # Skip items with missing question or passages

        for positive_passage in positive_passages:
            # Add the positive example
            processed_examples.append({
                'question': question,
                'passage': positive_passage,
                'label': 1
            })

            # Generate a corresponding negative example
            negative_passage = random.choice(all_passages)
            # Ensure the negative passage is not one of the positive passages for this question
            while negative_passage in positive_passages:
                negative_passage = random.choice(all_passages)

            processed_examples.append({
                'question': question,
                'passage': negative_passage,
                'label': 0
            })

    # Shuffle the dataset for randomness
    random.shuffle(processed_examples)

    # Convert to Hugging Face Dataset
    return Dataset.from_list(processed_examples)


print("\n--- Preparing Datasets ---")
# Load the training and validation sets
train_dataset = load_data_with_negatives(train_path)
val_dataset = load_data_with_negatives(val_path)

if not train_dataset or not val_dataset:
    print("\nCritical error during data loading. Exiting script.")
    exit()

print(f"Training dataset size: {len(train_dataset)}")
print(f"Validation dataset size: {len(val_dataset)}")


# --- 3. Metrics Calculation ---

def compute_metrics(eval_pred):
    """
    Calculates and returns accuracy, F1, precision, and recall.
    """
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)

    accuracy = accuracy_score(labels, predictions)
    f1 = f1_score(labels, predictions, average='binary')
    precision = precision_score(labels, predictions, average='binary')
    recall = recall_score(labels, predictions, average='binary')

    return {
        'accuracy': accuracy,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }

# --- 4. Main Processing Loop ---

all_model_results = {}

print("\n--- Starting Model Fine-Tuning and Evaluation Loop ---")

for model_name, checkpoint in models_to_process.items():
    print(f"\n{'='*20} Processing Model: {model_name} ({checkpoint}) {'='*20}")

    model_save_path = os.path.join(models_output_folder, model_name)

    # --- 4.A. Fine-Tuning ---
    if os.path.exists(model_save_path):
        print(f"Fine-tuned model for '{model_name}' already exists. Skipping training.")
    else:
        print(f"Fine-tuning new model for '{model_name}'.")

        # Load pre-trained model and tokenizer
        tokenizer = AutoTokenizer.from_pretrained(checkpoint)
        model = AutoModelForSequenceClassification.from_pretrained(checkpoint, num_labels=2).to(device)

        # Tokenization function
        def tokenize_function(examples):
            # The tokenizer for sentence-pair tasks takes two sequences
            return tokenizer(
                examples['question'],
                examples['passage'],
                truncation=True,
                padding='max_length',
                max_length=256 # A reasonable max length
            )

        # Tokenize datasets
        print("Tokenizing datasets...")
        tokenized_train_dataset = train_dataset.map(tokenize_function, batched=True)
        tokenized_val_dataset = val_dataset.map(tokenize_function, batched=True)

        # For backward compatibility, we calculate steps per epoch to manually set
        # logging, evaluation, and saving intervals.
        train_batch_size = 16
        # Avoid division by zero if the dataset is empty
        steps_per_epoch = len(tokenized_train_dataset) // train_batch_size if len(tokenized_train_dataset) > 0 else 1

        # Configure training arguments for maximum backward compatibility.
        # We set `load_best_model_at_end=False` to bypass the validation error and
        # will manually load the best checkpoint after training.
        training_args = TrainingArguments(
            output_dir=os.path.join(models_output_folder, f"{model_name}-checkpoints"),
            num_train_epochs=4,
            per_device_train_batch_size=train_batch_size,
            per_device_eval_batch_size=16,
            warmup_steps=500,
            weight_decay=0.01,
            logging_dir=os.path.join(models_output_folder, f"{model_name}-logs"),

            logging_steps=steps_per_epoch,
            save_steps=steps_per_epoch,
            eval_steps=steps_per_epoch,

            save_total_limit=2, # Keep a couple of checkpoints
            load_best_model_at_end=False, # Set to False to avoid validation error
            metric_for_best_model="f1",   # Still track 'f1' to find the best checkpoint
            greater_is_better=True,
            report_to="none", # Explicitly disable wandb logging
        )

        # Initialize the Trainer
        trainer = Trainer(
            model=model,
            args=training_args,
            train_dataset=tokenized_train_dataset,
            eval_dataset=tokenized_val_dataset,
            compute_metrics=compute_metrics,
            # EarlyStoppingCallback removed to prevent AssertionError with older transformers versions
        )

        # Execute training
        print("Starting training...")
        trainer.train()
        print("Training finished.")

        # Manually save the best model
        # The trainer.state tracks the best checkpoint based on the metric_for_best_model.
        best_checkpoint_path = trainer.state.best_model_checkpoint
        if best_checkpoint_path and os.path.exists(best_checkpoint_path):
            print(f"Found best checkpoint at: {best_checkpoint_path}")
            print(f"Loading best model and saving to: {model_save_path}")
            # Load the best model and tokenizer from the best checkpoint
            best_model = AutoModelForSequenceClassification.from_pretrained(best_checkpoint_path)
            best_tokenizer = AutoTokenizer.from_pretrained(best_checkpoint_path)
            # Save them to the final, clean destination
            best_model.save_pretrained(model_save_path)
            best_tokenizer.save_pretrained(model_save_path)
        else:
            # Fallback in case no best checkpoint was found (e.g., training stopped on the first epoch)
            print("No best checkpoint found. Saving the last model state.")
            trainer.save_model(model_save_path)
            tokenizer.save_pretrained(model_save_path)


    # --- 4.B. Evaluation ---
    print(f"Evaluating fine-tuned model: {model_name}")

    # Load the fine-tuned model and tokenizer from local disk
    eval_tokenizer = AutoTokenizer.from_pretrained(model_save_path)
    eval_model = AutoModelForSequenceClassification.from_pretrained(model_save_path).to(device)

    # Tokenize the validation set for evaluation
    def eval_tokenize_function(examples):
        return eval_tokenizer(
            examples['question'],
            examples['passage'],
            truncation=True,
            padding='max_length',
            max_length=256
        )

    tokenized_eval_set = val_dataset.map(eval_tokenize_function, batched=True)

    # Initialize a new Trainer instance for evaluation only
    # No need for most training arguments here
    eval_trainer = Trainer(
        model=eval_model,
        args=TrainingArguments(output_dir=f"./{model_name}-eval_results", per_device_eval_batch_size=16, report_to="none"),
        eval_dataset=tokenized_eval_set,
        compute_metrics=compute_metrics,
    )

    # Run evaluation
    evaluation_metrics = eval_trainer.evaluate()

    # Store and print the results
    all_model_results[model_name] = evaluation_metrics

    print(f"\n--- Results for {model_name} ---")
    for metric, value in evaluation_metrics.items():
        print(f"{metric}: {value:.4f}")
    print("-" * (20 + len(model_name) + len(" Results for ")))


# --- 5. Final Reporting ---
print(f"\n--- All Model Evaluations Complete ---")
print("Saving final results to:", results_output_path)

# Save the master dictionary of results to a JSON file
with open(results_output_path, 'w', encoding='utf-8') as f:
    json.dump(all_model_results, f, indent=4)

print("\nScript finished successfully.")
print("Final results dictionary:")
print(json.dumps(all_model_results, indent=2))

allDone()
```

#Passage Embedding

```python
import os
import json
import numpy as np
import pickle
import torch
import networkx as nx
from sentence_transformers import SentenceTransformer
from tqdm import tqdm

# --- Configuration ---
graph_path = "/content/drive/MyDrive/Colab Notebooks/modeldatalocation/graph.gpickle"
output_folder = "/content/drive/MyDrive/Colab Notebooks/modeldatalocation"
embedding_model_name = "sentence-transformers/all-mpnet-base-v2"
embedding_output_path = os.path.join(output_folder, "passage_embeddings.npy")
id_output_path = os.path.join(output_folder, "passage_ids.json")

# --- Load Graph ---
with open(graph_path, "rb") as f:
    G = pickle.load(f)

# --- Detect Device ---
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# --- Load Embedding Model ---
model = SentenceTransformer(embedding_model_name, device=device)

# --- Context Builder ---
def build_enriched_passage_text(G, passage_node_id):
    base_text = G.nodes[passage_node_id].get("text", "")
    context_snippets = []

    for neighbor in G.neighbors(passage_node_id):
        edge_type = G[passage_node_id][neighbor].get("type")
        node_data = G.nodes[neighbor]

        if node_data["type"] == "NamedEntity":
            snippet = f"[Entity] {node_data['term']}: {node_data.get('description', '')}"
            context_snippets.append(snippet)
        elif node_data["type"] == "DefinedTerm":
            snippet = f"[DefinedTerm] {node_data['term']}: {node_data.get('description', '')}"
            context_snippets.append(snippet)
        elif node_data["type"] == "Passage":
            snippet = f"[Related Passage] {node_data.get('text', '')}"
            context_snippets.append(snippet)

    full_context = base_text + "\n\n" + "\n".join(context_snippets)
    return full_context.strip()

# --- Collect and Encode Passages ---
passage_ids = []
passage_texts = []

for node_id, data in G.nodes(data=True):
    if data.get("type") == "Passage":
        enriched_text = build_enriched_passage_text(G, node_id)
        passage_ids.append(node_id)
        passage_texts.append(enriched_text)

print(f"Encoding {len(passage_texts)} passages...")

embeddings = model.encode(
    passage_texts,
    batch_size=32,
    show_progress_bar=True,
    convert_to_numpy=True
)

# --- Save Embeddings and IDs ---
np.save(embedding_output_path, embeddings)

with open(id_output_path, 'w', encoding='utf-8') as f:
    json.dump(passage_ids, f, indent=2)

print("\n✅ Embedding complete.")
print(f"Total Passages: {len(passage_ids)}")
print(f"Embedding Dimension: {embeddings.shape[1]}")
print(f"Saved Embeddings to: {embedding_output_path}")
print(f"Saved Passage IDs to: {id_output_path}")
```

# Retrieval + Reranking

```python
import os
import json
import torch
import pickle
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from sentence_transformers import SentenceTransformer, util

# --- Configuration ---
data_folder = "/content/drive/MyDrive/Colab Notebooks/modeldatalocation"
embedding_path = os.path.join(data_folder, "passage_embeddings.npy")
id_path = os.path.join(data_folder, "passage_ids.json")
graph_path = os.path.join(data_folder, "graph.gpickle")
test_set_path = os.path.join(data_folder, "ObliQA_MultiPassage_test.json")
results_output_folder = os.path.join(data_folder, "retrieval_reranking_results_full")
os.makedirs(results_output_folder, exist_ok=True)

# Fine-tuned model paths
fine_tuned_models = {
    "MiniLM": os.path.join(data_folder, "fine_tuned_models", "MiniLM"),
    "MPNet": os.path.join(data_folder, "fine_tuned_models", "MPNet"),
    "MSMarco": os.path.join(data_folder, "fine_tuned_models", "MSMarco"),
    "BERT": os.path.join(data_folder, "fine_tuned_models", "BERT"),
}

embedding_model_name = "sentence-transformers/all-mpnet-base-v2"
device = "cuda" if torch.cuda.is_available() else "cpu"
embedding_model = SentenceTransformer(embedding_model_name, device=device)

# Load graph and embeddings
with open(graph_path, "rb") as f:
    G = pickle.load(f)
with open(id_path, "r", encoding="utf-8") as f:
    passage_uids = json.load(f)
embeddings = np.load(embedding_path)

# Load test set
with open(test_set_path, "r", encoding="utf-8") as f:
    test_data = json.load(f)

# Retrieval + Reranking
all_results = {}
for q in test_data:
    qid = q["QuestionID"]
    question = q["Question"]

    # Step 1: Retrieve top-20 passages using cosine similarity
    query_embedding = embedding_model.encode(question, convert_to_tensor=True)
    cos_scores = util.pytorch_cos_sim(query_embedding, torch.tensor(embeddings).to(device))[0]
    top_k = min(20, len(passage_uids))
    top_indices = torch.topk(cos_scores, k=top_k).indices.tolist()

    retrieved_passages = []
    for idx in top_indices:
        uid = passage_uids[idx]
        node = G.nodes[uid]
        passage_id = node.get("passage_id", "")
        document_id = node.get("document_id", "")
        combined_uid = f"{document_id}|||{passage_id}"

        retrieved_passages.append({
            "uid": combined_uid,
            "passage_id": passage_id,
            "document_id": document_id,
            "text": node["text"]
        })

    all_results[qid] = {"question": question, "reranked": {}}

    # Step 2: Rerank with each fine-tuned model
    for model_name, model_path in fine_tuned_models.items():
        tokenizer = AutoTokenizer.from_pretrained(model_path)
        model = AutoModelForSequenceClassification.from_pretrained(model_path).to(device)
        model.eval()

        reranked = []
        for passage in retrieved_passages:
            inputs = tokenizer(
                question,
                passage["text"],
                return_tensors="pt",
                truncation=True,
                padding="max_length",
                max_length=256
            ).to(device)
            with torch.no_grad():
                logits = model(**inputs).logits
                score = torch.softmax(logits, dim=1)[0][1].item()
            reranked.append({
                "uid": passage["uid"],
                "passage_id": passage["passage_id"],
                "document_id": passage["document_id"],
                "text": passage["text"],
                "relevance_score": score
            })

        reranked = sorted(reranked, key=lambda x: x["relevance_score"], reverse=True)
        all_results[qid]["reranked"][model_name] = reranked

# Save results per model
for model_name in fine_tuned_models.keys():
    model_output_path = os.path.join(results_output_folder, f"{model_name}_reranked_results_full.json")
    model_results = {
        qid: data["reranked"][model_name]
        for qid, data in all_results.items()
    }
    with open(model_output_path, "w", encoding="utf-8") as f:
        json.dump(model_results, f, indent=2)

print("✅ Retrieval + Reranking results saved using UID format.")
```

```python
import json
import os
import pytrec_eval
import pandas as pd

# --- File Paths ---
data_folder = "/content/drive/MyDrive/Colab Notebooks/modeldatalocation"
qrel_path = os.path.join(data_folder, "qrels.trec")
results_folder = os.path.join(data_folder, "retrieval_reranking_results_full")

# --- Load QRELs ---
qrel = {}
with open(qrel_path, "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) >= 4:
            qid, _, uid, rel = parts[0], parts[1], " ".join(parts[2:-1]), parts[-1]
            qrel.setdefault(qid, {})[uid] = int(rel)

# --- Prepare PyTREC evaluator ---
evaluator = pytrec_eval.RelevanceEvaluator(qrel, {"recall", "map_cut", "ndcg_cut"})

# --- Collect Results for Each Model ---
model_scores = []

for filename in os.listdir(results_folder):
    if filename.endswith("_reranked_results_full.json"):
        model_name = filename.replace("_reranked_results_full.json", "")
        model_path = os.path.join(results_folder, filename)

        # Load run
        with open(model_path, "r", encoding="utf-8") as f:
            ranking_data = json.load(f)

        run = {}
        for qid, passages in ranking_data.items():
            run[qid] = {}
            for idx, passage in enumerate(passages):
                uid = passage["uid"]
                score = passage.get("relevance_score", 1.0 / (idx + 1))
                run[qid][uid] = float(score)

        # Evaluate
        results = evaluator.evaluate(run)

        # Aggregate
        R, MAP, NDCG = 0.0, 0.0, 0.0
        for scores in results.values():
            R += scores.get("recall_10", 0.0)
            MAP += scores.get("map_cut_10", 0.0)
            NDCG += scores.get("ndcg_cut_10", 0.0)

        n = len(results)
        model_scores.append({
            "Model": model_name,
            "Recall@10": R / n,
            "MAP@10": MAP / n,
            "nDCG@10": NDCG / n
        })

# --- Display Results ---
df = pd.DataFrame(model_scores).sort_values(by="Recall@10", ascending=False)
print("📊 Evaluation Summary:")
print(df.to_string(index=False, float_format="%.4f"))

# Optional: Save to CSV
output_csv = os.path.join(data_folder, "evaluation_summary.csv")
df.to_csv(output_csv, index=False)
print(f"\n📁 Results saved to: {output_csv}")
```
